```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def forward(x):
    return x * w
```

```python
def loss(x,y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
w = 1.0
w_list = []
mse_list = []
```

```python

for w in np.arange(0.0, 4.1, 0.1):
    print("w=", w)
    l_sum = 0
    for x_val, y_val in zip (x_data, y_data):
        y_pred_val = forward(x_val)
        l = loss(x_val, y_val)
        l_sum += l
        print("\t", x_val, y_val, y_pred_val, l)
    print("MSE=", l_sum/3)
    w_list.append(w)
    mse_list.append(l_sum/3)
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
        
        
```

```python
import torch
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
w = 1.0
```

```python
def forward(x):
    return w * x
```

```python
def loss(x,y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)
```

```python
def gradient(x,y):
    return 2 * x * (x * w - y)
```

```python
print("predict (before training)", 4, forward(4))
```

```python
for epoch in range (100):
    for x_val, y_val in zip(x_data, y_data):
        grad = gradient(x_val, y_val)
        w = w - 0.01 * grad
        print("\tgrad: ", x_val, y_val, grad)
        l = loss(x_val, y_val)
    print("Progress: ", epoch, "w=", w, "loss=", l)
print("predict (after training)", "4 hours", forward(4))
```
